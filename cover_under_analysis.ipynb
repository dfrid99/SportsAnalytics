```python
import pandas as pd

spreads_df = pd.read_csv('nba_betting_spread.csv')
totals_df = pd.read_csv('nba_betting_totals.csv')
games_df = pd.read_csv('nba_games_all.csv')
```

```python
spreads_df = spreads_df[spreads_df['book_name']=='Pinnacle Sports'].reset_index()
totals_df = totals_df[totals_df['book_name']=='Pinnacle Sports'].reset_index()

df = pd.merge(spreads_df, totals_df[['game_id', 'total1', 'total2',	'price1', 'price2']],
               on="game_id", how='left').reset_index()

print(df['spread1'].mean(), df['spread2'].mean())
```

```python
def get_game_stats(row):
    game = games_df[games_df['game_id']==row['game_id']].reset_index()
    game.set_index('team_id', inplace=True)
    if len(game) > 0:
        print(game.at[row['a_team_id'], 'is_home'])
        row['PTS_away'] = game.at[row['team_id'], 'pts']
        row['PTS_home'] = game.at[row['a_team_id'], 'pts']
    return row

df = df.apply(get_game_stats, axis=1)
df
```

```python
df['PTS_away'].isna().sum()
df['Total_real'] = df['PTS_away'] + df['PTS_home']
df['spread_real'] = df['PTS_home'] - df['PTS_away']
def get_over_spread(row):
    if row['Total_real'] > row['total1']:
        row['over'] = 1
    elif row['Total_real'] == row['total1']:
        row['over'] = 0
    else:
        row['over'] = -1
    
    spread_real = row['PTS_away'] - row['PTS_home']
    if row['spread2'] < 0:
        if spread_real < row['spread2']:
            row['fav_cover'] = 1
        elif spread_real > row['spread2']:
            row['fav_cover'] = -1
        else:
            row['fav_cover'] = 0
    else:
        if spread_real > row['spread2']:
            row['fav_cover'] = 1
        elif spread_real < row['spread2']:
            row['fav_cover'] = -1
        else:
            row['fav_cover'] = 0
    return row



df = df.apply(get_over_spread,axis=1)

count_cover = df[df['fav_cover']==1].count()['fav_cover']
count_nocov = df[df['fav_cover']==-1].count()['fav_cover']

print(count_nocov/(count_cover+count_nocov))
```

```python
big_spread = df[(abs(df['spread2'])>10) &
                (df['over'] != 0)&
                (df['fav_cover'] != 0)]

def get_parlay(row):
    row['over/cover'] = 0
    if (row['over'] == 1) and (row['fav_cover']== 1):
        row['over/cover'] = 1
    elif (row['over'] == -1) and (row['fav_cover']== -1):
        row['over/cover'] = -1
    return row

big_spread = big_spread.apply(get_parlay, axis=1)
big_spread
    
```

```python
count_over_cover = big_spread[big_spread['over/cover']==1].count()['over/cover']
count_under_nocover = big_spread[big_spread['over/cover']==-1].count()['over/cover']

print('fav cover/over', count_over_cover/len(big_spread))
print('underdog cover/under',count_under_nocover/len(big_spread))
```

```python
from statsmodels.tools.eval_measures import rmse
from sklearn.metrics import mean_squared_error, r2_score

rmse = rmse(df['spread_real'], df['spread1'],)
print(r2_score(df['spread_real'], df['spread1']))
print(rmse)
```
